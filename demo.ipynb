```python
import wfdb
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
%config InlineBackend.figure_format = "retina"
```

## Plot WFDB record

### Default

```python
# Read WFDB record
rec_path = "./sample-data/s40000306/40000306"
rd_record = wfdb.rdrecord(rec_path)

# Plot with default styles
wfdb.plot_wfdb(
    record=rd_record,
    time_units="samples",
    figsize=(20, 15),
    ecg_grids="all",
)
```

### Custom

```python
# Read WFDB record
rec_path = "./sample-data/s40000306/40000306"
rd_record = wfdb.rdrecord(rec_path)

# Plot with custom styles
fig = wfdb.plot_wfdb(
    record=rd_record,
    time_units="samples",
    figsize=(20, 15),
    ecg_grids="",      # Remove grid
    title=" ",         # Empty title
    close_axis=True,   # Hide axis
    line_color="w",    # White lines
    bg_color="k",      # Black background
    return_fig=True,   # Return the figure object
)

# Save figure
fig.savefig("./output/ecg_s40000306.jpg", bbox_inches='tight', pad_inches=0)
```
